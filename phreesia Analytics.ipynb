# Most visited client by month

```python
df.groupby("Visit Month")["Visit Count"].count()
```

# Most Visited speciality

```python
splty=df.groupby("Client Specialty").agg(
    most_visited_speciality=("Visit Count","sum")
).reset_index()
splty
splty.sort_values(by="most_visited_speciality",ascending=False)
```

```python
sns.lineplot(data=splty,x=splty["Client Specialty"],y=splty["most_visited_speciality"],marker="o")
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
